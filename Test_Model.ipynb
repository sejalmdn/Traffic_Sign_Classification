# Testing on local image

```python
import os
import cv2
import numpy as np
from keras.models import load_model
import matplotlib.pyplot as plt
from sign_names import labelid_to_name
```

```python
#load the model
model = load_model(os.path.join(os.getcwd(), "traffic_sign_model.h5"))
```

```python
#load the image
img_path = os.path.join(os.getcwd(), "test_img.jpg")
img = cv2.imread(img_path)

if img is None:
    raise ValueError("Failed to load the image.")

#preprocess the image
img = cv2.resize(img, (30, 30))
img = img/255
img = np.expand_dims(img, axis=0)
```

```python
#predict the label
pred = np.argmax(model.predict(img))
sign_name = labelid_to_name[pred]
plt.imshow(cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB))
plt.title(f"Predicted label: {sign_name} ({pred})")
plt.axis('off')
plt.show()
```
